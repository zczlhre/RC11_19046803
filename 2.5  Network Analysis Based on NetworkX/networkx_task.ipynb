```python
import networkx as nx
import csv
import matplotlib.pyplot as plt
```

```python
class Video_Game:
    """simulate a video game"""
    
    def __init__(self, name, platform, genre, publisher):
        """Initialize the properties of the video game"""
        self.name = name
        self.platform = platform
        self.genre = genre
        self.publisher = publisher
```

```python
filename = "video_game_sales.csv"
with open(filename) as f:
    reader = csv.reader(f)
    print(next(reader))
    print(next(reader))
```

```python
# Know how many lines are in the file
filename = "video_game_sales.csv"
with open(filename) as f:
    reader = csv.reader(f)
    rows = 0
    for r in reader:
        rows +=1
    print(rows)
```

```python
filename = "video_game_sales.csv"
with open(filename) as f:
    reader = csv.DictReader(f)
    
    # Create a collection of video games
    Video_Games= []
    for row in reader:
        name = row['name']
        platform = row['platform']
        genre = row['genre'] 
        publisher = row['genre'] 
        Video_Games.append(Video_Game(name, platform, genre, publisher))
```

```python
print(Video_Games[0].name)
print(Video_Games[0].platform)
```

```python
# Graph drawing, undirected graph
VG_Graph = nx.Graph()
```

```python
VG_Subset = Video_Games[0:100]
```

```python
for v in VG_Subset:
    VG_Graph.add_node(v,name = v.name[:15])
```

```python
for v1 in VG_Subset:
    for v2 in VG_Subset:
        if not v1 == v2:
            if v1.platform == v2.platform:
                 VG_Graph.add_edge(v1,v2)
            elif v1.genre == v2.genre:
                VG_Graph.add_edge(v1,v2)                  
print(VG_Graph)
```

```python
labels = nx.get_node_attributes(VG_Graph, 'name')
print(labels)
```

```python
options = {'node_color':'purple', 'node_size':300, 'edge_color':'black', 'width':0.3, 'labels':labels, 'with_labels':True}
plt.figure(1,figsize =(30,30))
nx.draw(VG_Graph, **options)
```

The graph shows the connections between Top 100 selling video games in the world based on their platform, function and genre.
