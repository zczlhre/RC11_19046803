```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams['font.sans-serif'] = ['SimHei']  
plt.rcParams['axes.unicode_minus'] = False  
plt.rcParams['figure.figsize'] = (20.0, 18.0)
```

```python
excel_file='data/test1.xlsx'
df = pd.read_excel(excel_file, 'Sheet1', index_col=None, na_values=['NA'])
df['building area classification']=df['wlc_building area']//50000
df['sunhours_avg']=df['sunhours']/df['wlc_building surface area']
df['radiation_avg']=df['radiation']/df['wlc_building surface area']
```

# Visualize statistical relationships
Statistical analysis is the process of understanding how variables in a data set relate to each other and how those relationships depend on other variables. Visualization can be a core component of this process because the human visual system can see trends and patterns in indicative relationships when data is properly visualized.

```python
sns.relplot(x="radiation", y="sunhours", hue="wlc_building area",palette="ch:r=-.5,l=.55",data=df)
plt.savefig('test.png', dpi = 300)
```

```python
sns.relplot(x="wlc_building volume factor", y="sunhours_avg",hue="radiation_avg", col="building area classification", col_wrap=3, data=df)
plt.savefig('test1.png', dpi = 300)
```
